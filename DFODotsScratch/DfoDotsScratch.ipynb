```python


import pandas as pd
from matplotlib import pyplot as plt

# read in CSV
df = pd.read_csv("dots.csv")
# df.sort_values("dot_index", inplace=True)
df.head()
```

```python
marker_map = {
    "Sea: sea age": "o",
    "Edge: otolith edge": "x",
}
plt.figure(figsize=(10,10))
for dot_type in df["dot_type"].unique():
    subdf = df.where(df["dot_type"] == dot_type).dropna()
    plt.scatter(subdf["x"], subdf["y"], marker=marker_map[dot_type], label=dot_type)
    plt.legend()
plt.plot(df["x"], df["y"], linestyle=":")
for index, row in df.iterrows():
    plt.annotate(row["dot_index"], xy = (row["x"]+5, row["y"]-5))
```

### but if I sort my DF by `dot_index` field...



```python
df.sort_values("dot_index", inplace=True)
plt.figure(figsize=(10,10))
for dot_type in df["dot_type"].unique():
    subdf = df.where(df["dot_type"] == dot_type).dropna()
    plt.scatter(subdf["x"], subdf["y"], marker=marker_map[dot_type], label=dot_type)
    plt.legend()

plt.plot(df["x"], df["y"], linestyle=":")

for index, row in df.iterrows():
    plt.annotate(row["dot_index"], xy = (row["x"]+5, row["y"]-5))
```
